### **Charts**

#### **Total Sales By Fat Content**

#### **Total Sales By Fat Content**

```python
total_sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
# print(tsf.index)

plt.pie(total_sales_by_fat,labels=total_sales_by_fat.index,autopct = '%.1f%%',startangle=90)
plt.axis('equal')
plt.title('Total Sales By Fat Content')
plt.show()
```

#### **Total Sales By Item Types**

```python
total_sales_by_item_type = df.groupby('Item Type')['Sales'].sum()

plt.figure(figsize=(10,8))
bars = plt.bar(total_sales_by_item_type.index,total_sales_by_item_type.values)
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.xticks(rotation=-90)
plt.title('Total Sales By Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,bar.get_height(),
         f'{bar.get_height():,.1f}',ha='center',va='bottom',fontsize=8)
    
plt.tight_layout()
plt.show()


```

#### **Fat Content By Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type','Item Fat Content']) ['Sales'].sum().unstack()
# grouped = grouped[['Regular','Low Fat']]
# print(grouped)

grouped.plot(kind='bar',figsize=(10,8),title='Fat Content By Outlet for Total Sales')
plt.xlabel('Outlet Location Type')
plt.ylabel('Sales')
plt.legend(title='Item Fat Content')
```

#### **Total Sales By Outlet Establishment**

```python
df['Outlet Establishment Year'].unique()
```

```python
tsoe = df.groupby('Outlet Establishment Year')['Sales'].sum()
# print(tsoe)
plt.figure(figsize=(10,6))
plt.plot( tsoe.index,tsoe.values,marker='o',linestyle='-')
plt.xticks(tsoe.index,rotation=45)
plt.xlabel('Establishment Year')
plt.ylabel('Sales')
plt.title('Total Sales By Outlet Establishment')

plt.tight_layout()
plt.show()
```

#### **Percentage of Sales By Outlet Size**


```python
sales_by_outlet_size = df.groupby('Outlet Size')['Sales'].sum()

plt.pie(sales_by_outlet_size,labels=sales_by_outlet_size.index,autopct='%1.1f%%',startangle=90)
plt.tight_layout()
plt.title('Sales By Outlet Size')
plt.show()
```

#### **Sales By Outlet Location**


```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales',ascending=False)

plt.figure(figsize=(8,3))
ax = sns.barplot(x='Sales',y='Outlet Location Type',data=sales_by_location)
plt.title('Sales By Outlet Location Type')
plt.xlabel('Sales')
plt.ylabel('Outlet Location Type')
plt.tight_layout()
plt.show()
```
